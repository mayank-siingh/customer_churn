```python
import pandas as pd
xls = pd.ExcelFile('e commerce data-1.xlsx')
df1 = pd.read_excel(xls, 'Sheet1')
df2 = pd.read_excel(xls, 'Sheet2')
```

```python
df1
```

```python
df2
```

Finding Interesting columns from each of the sheet to work upon finding correlation

```python
dff1 = df1[['Customer id','Order ID','Issue Category','Issue Type','Priority']]
```

```python
dff2 = df2[['Customer ID','OrderId','gender','City','ProdCategory','PaymentStatus','OrderAmt','Exclusive/Non-Exclusive']]
```

Now merging the 2 dataframes on CUSTOMER ID's and ORDER ID's, but before that renaming the columns as 'Customer_id' and 'Order_id' in both dataframes

```python
dff1.rename(columns={"Customer id": "Customer_id", "Order ID": "Order_id"},inplace=True)
dff2.rename(columns={"Customer ID": "Customer_id", "OrderId": "Order_id"},inplace=True)
```

```python
dff = pd.merge(dff1,dff2,on=['Customer_id','Order_id'])
```

```python
dff
```

Making a list of unique customers of dff

```python
unique = pd.DataFrame(list(dff['Customer_id'].unique()),columns=['Customer_id'])
```

Importing final dataset

```python
df_final = pd.read_csv('last.csv')
```

```python
df_final
```

```python
m = []
for i in range(len(unique)):
    for j in range(len(df_final)):
        if unique['Customer_id'][i] == df_final['Customer_id'][j]:
            m.append(df_final['Churn_value'][j])
```

```python
unique['Churn_value'] = m
```

Now merging 'unique' df and 'dff' df so that we have everything to corelated in dff

```python
DF = pd.merge(unique,dff,on='Customer_id')
```

```python
DF
```

Now looking at crosstabs frequencies of 2 groups i.e 'Issue Category' and 'Priority'

```python
crosstab = pd.crosstab(DF['Priority'], DF['Issue Category'])
crosstab
```

Chi-square Test of Independence

```python
from scipy import stats
stats.chi2_contingency(crosstab)
```

```python
import researchpy as rp
table, results = rp.crosstab(DF['Priority'], DF['Issue Category'], prop= 'col', test= 'chi-square')  
table
```

```python
results
```

As the p-value is < 0.05, then Null hypothesis can be said as wrong. And there must be correlation between the columns.
Also for Cramer's V value very close to 1, there is strong positive correlation between the two columns
